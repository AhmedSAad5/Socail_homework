```python
                                                  #==> direct graph
import networkx as nx
%matplotlib inline
```

```python
A = nx.DiGraph()
A.add_edges_from([(1,2),(2,3),(2,4),(3,2),(4,1),(6,3),(6,1)])
A.add_node(5)
nx.draw(A, with_labels=True, font_color='black', node_size=1000,node_color='red')
```

```python
Adj = nx.adjacency_matrix(A).todense()
Adj
```

```python
from collections import Counter
import matplotlib.pyplot as plt

degree_sequence = [A.degree(i) for i in A.nodes]
degree_counts = Counter(degree_sequence)
min_degree, max_degree = min(degree_counts.keys()), max(degree_counts.keys())
plt.xlabel("Degree", fontsize=18)
plt.ylabel("Number of Nodes", fontsize=14)
plot_x = list(range(min_degree, max_degree + 1))
plot_y = [degree_counts.get(x, 0) for x in plot_x]
plt.bar(plot_x, plot_y)
```

```python
def plot_degree_dist(A):
    
    degrees = A.degree()
    degrees = dict(degrees)
    values = sorted(set(degrees.values()))
    print(values)
    histo = [list(degrees.values()).count(x) for x in values]
    P_k = [x / A.order() for x in histo]
    print(len(P_k))
    
    plt.figure()
    plt.bar(values, P_k)
    plt.xlabel("k",fontsize=22)
    plt.ylabel("p(k)", fontsize=22)
    plt.title("Degree Distribution", fontsize=20)
    
    plt.show()
    plt.figure()
    plt.grid(False)
    plt.loglog(values, P_k, "bo")
    plt.xlabel("k", fontsize=20)
    plt.ylabel("log p(k)", fontsize=20)
    plt.title("log Degree Distribution")
    plt.show()
    plt.show()
```

```python
plot_degree_dist(A)
```

```python
                                    #==> undirect graph
import networkx as nx
%matplotlib inline
```

```python
G = nx.Graph()
G.add_nodes_from([1,2,3,4,5,6,7])
G.add_edges_from([
    (1,2),
    (1,3),
    (1,4),
    (1,6),
    (2,4),
    (2,3),
    (6,3),
    (3,4),
])
nx.draw(G, with_labels=True, font_color='red', node_size=1000,node_color='red')
```

```python
Adj = nx.adjacency_matrix(G).todense()
Adj
```

```python
from collections import Counter
import matplotlib.pyplot as plt

degree_sequence = [G.degree(n) for n in G.nodes]
degree_counts = Counter(degree_sequence)
min_degree, max_degree = min(degree_counts.keys()), max(degree_counts.keys())
plt.xlabel("Degree", fontsize=15)
plt.ylabel("Number of Nodes", fontsize=15)
plot_x = list(range(min_degree, max_degree + 1))
plot_y = [degree_counts.get(x, 0) for x in plot_x]
plt.bar(plot_x, plot_y)
```

```python
def plot_degree_dist(G):
    
    degrees = G.degree()
    degrees = dict(degrees)
    values = sorted(set(degrees.values()))
    print(values)
    histo = [list(degrees.values()).count(x) for x in values]
    P_k = [x / G.order() for x in histo]
    print(len(P_k))
    
    plt.figure()
    plt.bar(values, P_k)
    plt.xlabel("k",fontsize=20)
    plt.ylabel("p(k)", fontsize=20)
    plt.title("Degree Distribution", fontsize=20)
    
    plt.show()
    plt.figure()
    plt.grid(False)
    plt.loglog(values, P_k, "bo")
    plt.xlabel("k", fontsize=20)
    plt.ylabel("log p(k)", fontsize=20)
    plt.title("log Degree Distribution")
    plt.show()
    plt.show()  
```
